```python
import sys, os
from gensim.models import Word2Vec
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from matplotlib import pyplot
import tqdm as tqdm
from gensim.models import TranslationMatrix
model = Word2Vec.load('embeddings/k_model_128_30_50.pkl')
words=[]



```

```python
# 基于2d PCA拟合数据
X = model.wv[model.wv.index_to_key]
pca = PCA(n_components=2)
result = pca.fit_transform(X)
#result = TSNE(n_components=2).fit_transform(result_pca)
# 可视化展示
pyplot.xticks([]) 
pyplot.yticks([]) 
pyplot.rcParams['font.size'] = 8  # 字体大小
pyplot.rcParams['axes.unicode_minus'] = False  # 正常显示负号
pyplot.figure(figsize=(20,20))
pyplot.scatter(result[:, 0], result[:, 1],color='#86C166')
ad = list(model.wv.index_to_key)
words=[]
for i in range(len(ad)):
	try:
		words.append(int(ad[i]))
	except ValueError:
		words.append(' ')

for i, word in enumerate(words):
	pyplot.annotate(word, xy=(result[i, 0], result[i, 1]))
pyplot.show()
```
